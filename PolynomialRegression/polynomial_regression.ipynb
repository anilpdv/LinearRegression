```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
#importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1].values.reshape(-1,1)
y = dataset.iloc[:,-1].values
```

```python
#ploting a sctter plot between the level and salary
plt.style.use('ggplot')
plt.scatter(dataset['Level'].values,dataset['Salary'].values,c='r',edgecolors='b',alpha=0.6)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Level vs Salary')
plt.show()
```

```python
#fitting linear regression to the dataset
from sklearn.linear_model import LinearRegression

#creating an object of the linear regression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
lin_pred = list(lin_reg.predict(X))
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(13,6))
plt.scatter(dataset['Level'].values,dataset['Salary'].values,c='r',edgecolors='b',alpha=0.6)
plt.xlim(0,11)
plt.plot(X,lin_pred,color='r',linewidth=2)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Level vs Salary')
plt.show()
```

```python
#fitting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures

#transform the matrix to polynomial terms
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)

#fitting the linear regression model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
#predicting the model 
poly_predict = lin_reg_2.predict(X_poly)
```

```python
#ploting the fitting curve
plt.style.use('ggplot')
plt.figure(figsize=(13,6))
plt.scatter(dataset['Level'].values,dataset['Salary'].values,c='k')
plt.plot(X,poly_predict,color='c')
#plt.plot(lin_pred,color='b',linewidth=2)
plt.xlim(0,11)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Level vs Salary')
plt.show()
```

```python
#predicting a new result with linear regression
lin_reg.predict(6.5)
```

```python
#prediction a new result with polynomial regression
lin_reg_2.predict(poly_reg.fit_transform(6.5))
```
