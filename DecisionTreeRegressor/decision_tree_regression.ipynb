```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,-1].values
```

```python
#ploting a sctter plot between the level and salary
plt.style.use('ggplot')
plt.scatter(dataset['Level'].values,dataset['Salary'].values,c='r',edgecolors='b',alpha=0.6)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Level vs Salary')
plt.show()
```

```python
#fitting the decision tree regressor to the dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

```python
regressor.predict(6.5)
```

```python
#ploting a sctter plot between the level and salary

plt.style.use('ggplot')
plt.scatter(dataset['Level'].values,dataset['Salary'].values,c='r',edgecolors='b',alpha=0.6)
plt.plot(X,regressor.predict(X))
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Level vs Salary')
plt.show()
```

```python
def rmse(pred,actual):
    count = len(actual)
    mse = (pred-actual)**2
    avg = sum(mse)/count
    rms = np.sqrt(avg)
    return rms
rmse(regressor.predict(X),y)
```

```python
X_grid = np.arange(min(X),max(X),0.01)
X_grid = X_grid.reshape(len(X_grid),1)
plt.style.use('ggplot')
plt.scatter(dataset['Level'].values,dataset['Salary'].values,c='r',edgecolors='b',alpha=0.6)
plt.plot(X_grid,regressor.predict(X_grid))
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Level vs Salary')
plt.show()
```
